```python
#FILTRO SOBEL
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar la imagen a color
imagen = cv2.imread('miche.jpg')

# Convertir la imagen a escala de grises
imagen_gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)

# Aplicar el filtro Sobel en dirección X (horizontal)
sobel_x = cv2.Sobel(imagen_gris, cv2.CV_64F, 1, 0, ksize=3)

# Aplicar el filtro Sobel en dirección Y (vertical)
sobel_y = cv2.Sobel(imagen_gris, cv2.CV_64F, 0, 1, ksize=3)

# Calcular la magnitud del gradiente
magnitud = np.sqrt(sobel_x**2 + sobel_y**2)

# Normalizar la magnitud para mostrarla en blanco y negro
magnitud_normalizada = cv2.normalize(magnitud, None, 0, 255, cv2.NORM_MINMAX, cv2.CV_8U)

# Guardar la imagen resultante en un archivo
cv2.imwrite('bordes_miche.jpg', magnitud_normalizada)

# Mostrar la imagen de características de bordes en blanco y negro
plt.imshow(magnitud_normalizada, cmap='gray')
plt.title('Características de Bordes en Blanco y Negro')
plt.axis('off')
plt.show()
```

```python
#FILTRO OTSU
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar la imagen en escala de grises
imagen = cv2.imread('bordes_miche.jpg', cv2.IMREAD_GRAYSCALE)

# Aplicar el umbral de Otsu
umbral, imagen_binaria = cv2.threshold(imagen, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Guardar la imagen binarizada
cv2.imwrite('otsu_miche.jpg', imagen_binaria)

# Mostrar la imagen original y la imagen binarizada resultante
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(imagen, cmap='gray')
plt.title('Imagen Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(imagen_binaria, cmap='gray')
plt.title(f'Imagen Binarizada (Umbral de Otsu = {umbral})')
plt.axis('off')

plt.show()
```

```python
# INVERTIR OTSU
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar la imagen binarizada (resultado del umbral de Otsu)
imagen_binaria = cv2.imread('otsu_miche.jpg', cv2.IMREAD_GRAYSCALE)

# Invertir la imagen binaria
imagen_invertida = cv2.bitwise_not(imagen_binaria)

# Guardar la imagen invertida
cv2.imwrite('invertida_otsu.jpg', imagen_invertida)

# Mostrar la imagen binaria original y la imagen invertida
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(imagen_binaria, cmap='gray')
plt.title('Imagen Binarizada Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(imagen_invertida, cmap='gray')
plt.title('Imagen Invertida')
plt.axis('off')

plt.show()
```

```python
#ESCALA DE GRISES
import cv2
import matplotlib.pyplot as plt

# Cargar la imagen "miche"
miche = cv2.imread('miche.jpg')

# Convertir "miche" a escala de grises
miche_gris = cv2.cvtColor(miche, cv2.COLOR_BGR2GRAY)

# Guardar la imagen en escala de grises
cv2.imwrite('esc_gri.jpg', miche_gris)

# Mostrar la imagen original y la imagen en escala de grises
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(miche, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original (Miche)')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(miche_gris, cmap='gray')
plt.title('Imagen en Escala de Grises')
plt.axis('off')

plt.show()
```

```python
#RESTA DE IMAGENES
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar la imagen "miche"
im1 = cv2.imread('miche.jpg')

# Cargar la imagen "invertida_otsu"
im2 = cv2.imread('invertida_otsu.jpg')

# Restar "invertida_otsu" de "miche"
resultado_resta = cv2.subtract( im1,im2)

# Mostrar las tres imágenes
plt.figure(figsize=(16, 6))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(im1, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original (Miche)')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(im2, cv2.COLOR_BGR2RGB))
plt.title('Imagen Invertida Otsu')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(resultado_resta, cv2.COLOR_BGR2RGB))
plt.title('Resultado de la Resta')
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar la imagen "miche"
im1 = cv2.imread('miche.jpg')

# Cargar la imagen "invertida_otsu"
im2 = cv2.imread('invertida_otsu.jpg')

# Calcular la diferencia absoluta entre "miche" e "invertida_otsu"
resultado_diferencia = cv2.absdiff(im1, im2)

# Mostrar las tres imágenes
plt.figure(figsize=(16, 6))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(im1, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original (Miche)')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(im2, cv2.COLOR_BGR2RGB))
plt.title('Imagen Invertida Otsu')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(resultado_diferencia, cv2.COLOR_BGR2RGB))
plt.title('Resultado de la Diferencia Absoluta')
plt.axis('off')

plt.show()
```
