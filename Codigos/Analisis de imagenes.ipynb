```python
import cv2
import matplotlib.pyplot as plt

def mostrar_histogramas(imagen, titulo):
    # Convertir de BGR a RGB para mostrar con Matplotlib
    imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

    # Crear la figura
    plt.figure(figsize=(12, 4))

    # Mostrar la imagen
    plt.subplot(1, 4, 1)
    plt.imshow(imagen_rgb)
    plt.title(titulo)
    plt.axis('off')

    # Mostrar histograma R
    plt.subplot(1, 4, 2)
    histograma_r = cv2.calcHist([imagen], [0], None, [256], [0, 256])
    plt.plot(histograma_r, color='r')
    plt.xlim([0, 256])
    plt.title('Histograma R')

    # Mostrar histograma G
    plt.subplot(1, 4, 3)
    histograma_g = cv2.calcHist([imagen], [1], None, [256], [0, 256])
    plt.plot(histograma_g, color='g')
    plt.xlim([0, 256])
    plt.title('Histograma G')

    # Mostrar histograma B
    plt.subplot(1, 4, 4)
    histograma_b = cv2.calcHist([imagen], [2], None, [256], [0, 256])
    plt.plot(histograma_b, color='b')
    plt.xlim([0, 256])
    plt.title('Histograma B')

    # Ajustar diseño
    plt.tight_layout()

    # Mostrar la figura
    plt.show()

# Leer las imágenes
imagen_lun1 = cv2.imread('37.jpg')
imagen_lun2 = cv2.imread('imagen31.jpg')
imagen_realzada = cv2.imread('imagen_realzada.jpg')

# Mostrar la primera imagen con sus histogramas
mostrar_histogramas(imagen_lun1, 'Imagen Celular')

# Mostrar la segunda imagen con sus histogramas
mostrar_histogramas(imagen_lun2, 'Imagen Modulo')

# Mostrar la tercera imagen con sus histogramas
mostrar_histogramas(imagen_realzada, 'Imagen Procesada')


```

```python
import cv2
import matplotlib.pyplot as plt

def mostrar_histogramas(imagen1, imagen2, imagen3):
    # Convertir de BGR a RGB para mostrar con Matplotlib
    imagen1_rgb = cv2.cvtColor(imagen1, cv2.COLOR_BGR2RGB)
    imagen2_rgb = cv2.cvtColor(imagen2, cv2.COLOR_BGR2RGB)
    imagen3_rgb = cv2.cvtColor(imagen3, cv2.COLOR_BGR2RGB)

    # Crear la figura
    plt.figure(figsize=(12, 4))

    # Mostrar histograma R
    plt.subplot(1, 3, 1)
    histograma_r_1 = cv2.calcHist([imagen1], [0], None, [256], [0, 256])
    histograma_r_2 = cv2.calcHist([imagen2], [0], None, [256], [0, 256])
    histograma_r_3 = cv2.calcHist([imagen3], [0], None, [256], [0, 256])
    plt.plot(histograma_r_1, color='r', label='Imagen Original')
    plt.plot(histograma_r_2, color='g', label='Imagen Modulo')
    plt.plot(histograma_r_3, color='b', label='Imagen Procesada')
    plt.xlim([0, 256])
    plt.title('Histograma R')
    plt.legend()

    # Mostrar histograma G
    plt.subplot(1, 3, 2)
    histograma_g_1 = cv2.calcHist([imagen1], [1], None, [256], [0, 256])
    histograma_g_2 = cv2.calcHist([imagen2], [1], None, [256], [0, 256])
    histograma_g_3 = cv2.calcHist([imagen3], [1], None, [256], [0, 256])
    plt.plot(histograma_g_1, color='r', label='Imagen Original')
    plt.plot(histograma_g_2, color='g', label='Imagen Modulo')
    plt.plot(histograma_g_3, color='b', label='Imagen Procesada')
    plt.xlim([0, 256])
    plt.title('Histograma G')
    plt.legend()

    # Mostrar histograma B
    plt.subplot(1, 3, 3)
    histograma_b_1 = cv2.calcHist([imagen1], [2], None, [256], [0, 256])
    histograma_b_2 = cv2.calcHist([imagen2], [2], None, [256], [0, 256])
    histograma_b_3 = cv2.calcHist([imagen3], [2], None, [256], [0, 256])
    plt.plot(histograma_b_1, color='r', label='Imagen Original')
    plt.plot(histograma_b_2, color='g', label='Imagen Modulo')
    plt.plot(histograma_b_3, color='b', label='Imagen Procesada')
    plt.xlim([0, 256])
    plt.title('Histograma B')
    plt.legend()

    # Ajustar diseño
    plt.tight_layout()

    # Mostrar la figura
    plt.show()

# Leer las imágenes
imagen_original = cv2.imread('37.jpg')
imagen_modulo = cv2.imread('imagen31.jpg')
imagen_procesada = cv2.imread('imagen_realzada.jpg')  # Cambiar el nombre de la imagen si es diferente

# Mostrar los histogramas de las tres imágenes en una sola gráfica
mostrar_histogramas(imagen_original, imagen_modulo, imagen_procesada)

```

```python
#Diferencia de Píxeles
import cv2
import matplotlib.pyplot as plt

# Cargar las imágenes
imagen10 = cv2.imread('imagen10.jpg')
imagen_realzada = cv2.imread('imagen_realzada.jpg')

# Diferencia de píxeles
diferencia = cv2.absdiff(imagen10, imagen_realzada)

# Mostrar las imágenes y la diferencia
plt.figure(figsize=(10, 5))

plt.subplot(131)
plt.imshow(cv2.cvtColor(imagen10, cv2.COLOR_BGR2RGB))
plt.title('Imagen 1')
plt.axis('off')

plt.subplot(132)
plt.imshow(cv2.cvtColor(imagen_realzada, cv2.COLOR_BGR2RGB))
plt.title('Imagen Realzada')
plt.axis('off')

plt.subplot(133)
plt.imshow(cv2.cvtColor(diferencia, cv2.COLOR_BGR2RGB))
plt.title('Diferencia de Píxeles')
plt.axis('off')

plt.show()
```

```python
# SSIM (Structural Similarity Index)
from skimage.metrics import structural_similarity as ssim
import cv2
import matplotlib.pyplot as plt

# Cargar las imágenes
imagen44 = cv2.imread('imagen10.jpg') 
imagen_realzada = cv2.imread('imagen_realzada.jpg')

# Convertir a escala de grises para SSIM
imagen44_gris = cv2.cvtColor(imagen44, cv2.COLOR_BGR2GRAY)
imagen_realzada_gris = cv2.cvtColor(imagen_realzada, cv2.COLOR_BGR2GRAY)

# Calcular SSIM
ssim_index, _ = ssim(imagen44_gris, imagen_realzada_gris, full=True)

# Mostrar las imágenes y el índice SSIM
plt.figure(figsize=(10, 5))

plt.subplot(131)
plt.imshow(imagen44)
plt.title('Imagen 1')  
plt.axis('off')

plt.subplot(132)
plt.imshow(imagen_realzada)
plt.title('Imagen Realzada')
plt.axis('off')

plt.subplot(133)
plt.text(0.5, 0.5, f'SSIM: {ssim_index:.2f}', fontsize=15, ha='center')
plt.axis('off')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Cargar las imágenes
imagen10 = cv2.imread('imagen10.jpg')
imagen_realzada = cv2.imread('imagen_realzada.jpg')

# Convertir a escala de grises para histograma 2D
imagen10_gris = cv2.cvtColor(imagen10, cv2.COLOR_BGR2GRAY)
imagen_realzada_gris = cv2.cvtColor(imagen_realzada, cv2.COLOR_BGR2GRAY)

# Calcular el histograma 2D
hist2d_10 = cv2.calcHist([imagen10], [0, 1], None, [256, 256], [0, 256, 0, 256])
hist2d_realzada = cv2.calcHist([imagen_realzada], [0, 1], None, [256, 256], [0, 256, 0, 256])

# Mostrar el histograma 2D
plt.figure(figsize=(12, 5))

plt.subplot(121)
plt.imshow(hist2d_10, interpolation='nearest')
plt.title('Histograma 2D - Imagen 1')
plt.xlabel('Canal B')
plt.ylabel('Canal G')
plt.colorbar()

plt.subplot(122)
plt.imshow(hist2d_realzada, interpolation='nearest')
plt.title('Histograma 2D - Imagen Realzada')
plt.xlabel('Canal B')
plt.ylabel('Canal G')
plt.colorbar()

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar las imágenes
imagen10 = cv2.imread('imagen10.jpg', cv2.IMREAD_GRAYSCALE)
imagen_realzada = cv2.imread('imagen_realzada.jpg', cv2.IMREAD_GRAYSCALE)

# Normalizar imágenes
imagen10_norm = cv2.normalize(imagen10, None, 0, 255, cv2.NORM_MINMAX)
imagen_realzada_norm = cv2.normalize(imagen_realzada, None, 0, 255, cv2.NORM_MINMAX)

# Calcular la Normalized Cross-Correlation
result = cv2.matchTemplate(imagen10_norm, imagen_realzada_norm, cv2.TM_CCOEFF_NORMED)
min_val, max_val, _, _ = cv2.minMaxLoc(result)

# Mostrar imágenes y resultados
plt.figure(figsize=(10, 5))

plt.subplot(221)
plt.imshow(imagen10, cmap='gray')
plt.title('Imagen 1')
plt.axis('off')

plt.subplot(222)
plt.imshow(imagen_realzada, cmap='gray')
plt.title('Imagen Realzada')
plt.axis('off')

plt.subplot(223)
plt.imshow(result, cmap='jet')
plt.title('Normalized Cross-Correlation')
plt.colorbar()

plt.subplot(224)
plt.text(0.5, 0.5, f'Max Value: {max_val:.2f}', fontsize=15, ha='center')
plt.axis('off')

plt.show()
```
